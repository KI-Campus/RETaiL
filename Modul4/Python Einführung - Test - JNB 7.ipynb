### Aufgabe

Der Code weiter unten erzeugt ein Diagramm. Verändere den Code so, dass das Diagramm wie dieses hier aussieht:



Nutze zur Hilfe gern die im Lücken-Code in Moodle vorgegebenen Auswahlmöglichkeiten.

Verändere insbesondere:

Fülle anschließend den Lücken-Code in Moodle aus.



```python
pivot_table = df.pivot_table(index='Eiscremesorte', columns='Verkäufer', values='Absatz', aggfunc='sum', fill_value=0)

# Sortieren der Pivot-Tabelle nach dem Absatz von Josef, absteigend
pivot_table = pivot_table.sort_values(by=('Klaus'), ascending=False)

# Farben definieren
colors = {'Josef': 'lightblue', 'Klaus': 'navy'}

# Erstellen eines Balkendiagramms mit unterschiedlichen Farben für jeden Verkäufer
ax = pivot_table.plot(kind='bar', color=[colors.get(x, '#333333') for x in pivot_table.columns], figsize=(6, 8))
plt.title('Absatz der Eiscremesorten nach Verkäufern')
plt.xlabel('XXX')
plt.ylabel('YYY')
plt.show()
```
