## Elbow Methode

```python
# Definieren eines Bereichs für die Anzahl der Cluster, die wir testen möchten 
k_rng = range(1, 10)

# Liste zur Speicherung des Sum of Squared Error (SSE) für jeden Wert von k
sse = []

# Für jeden Wert in k_rng ...
for k in k_rng:
    # Initialisieren des KMeans-Algorithmus mit der aktuellen Anzahl von Clustern (k)
    km = KMeans(n_clusters=k)
    
    # Fitting des KMeans-Algorithmus auf die Daten
    km.fit(df[['alter_t', 'nettoeinkommen_eur_t']])
    
    # Den SSE des aktuellen Modells zur Liste hinzufügen
    sse.append(km.inertia_)

# Ausgeben der berechneten SSE-Werte für jeden Clusterwert
print("SSE für verschiedene Clusterzahlen:", sse)
```

## Elbow Graph zeichnen

```python
# Visualisieren der SSE-Werte für verschiedene Clusterzahlen mit einem Liniendiagramm
plt.plot(k_rng, sse, marker='o', linestyle='-', color='b')

# Festlegen der Achsenbeschriftungen für bessere Verständlichkeit
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Error (SSE)')

# Hinzufügen eines Titels zum Diagramm
plt.title('Elbow Chart: SSE vs. Number of Clusters')

# Anzeigen des Liniendiagramms
plt.show()
```
