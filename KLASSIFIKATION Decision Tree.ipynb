# Decision Trees

## Bibliotheken einbinden und Daten vorbereiten

```python
# Bibliotheken einbinden
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Daten importieren
obst = pd.read_csv("obst.csv", sep=";")
```

```python
# Daten ausgeben
obst.head()
```

```python
# Daten in Feature-Matrix und Target-Vektor splitten
features = obst.iloc[:, 0:2]
target = obst.iloc[:, 2]
```

## Decision Tree trainieren und plotten 

```python
dt = tree.DecisionTreeClassifier(max_depth=1)
dt = dt.fit(features, target)
tree.plot_tree(dt)
```

## Daten plotten

```python
fg = sns.FacetGrid(data=obst, hue='Frucht')
fg.map(plt.scatter, 'Hoehe', 'Breite').add_legend()
```
